```python
import numpy as np
import numpy.matlib
```

####  The below code is used to generate the data points.

```python
# Generating two cluster od data points. The center are c1 and c2 respectively
# Class1 => Class1 datapoints
# Class2 => Class2 data points
#no => no of points in a class
c1 = [2,3]
c2 = [100,3]
no = 100
Class1 = np.matlib.repmat(c1, no,1) + np.random.randn(no,len(c1))
Class2 = np.matlib.repmat(c2, no,1) + np.random.randn(no,len(c2))
Data = np.append(Class1,Class2,axis = 0)
Trainlabel  = np.append(np.zeros((no,1)),np.ones((no,1)),axis = 0)
```

```python
(m,n) = Data.shape
print Data.shape
print Trainlabel.shape
#print Trainlabel
```

#### Defining the sigmoid function

```python
def sigmoid(x,play):
    z =  1/(1+np.exp(-x))
    if (play == "Forward"):
        return z
    elif (play == "Backward"):
        return z*(1-z)
```

#### Intializing the weights and bias to zeros

```python

learningrate = 0.01
numiter = 10000
```

```python
X = Data.T
print "Shape of X",X.shape
y = Trainlabel.T
w = np.zeros((1,n))
print "Shape ofy",y.shape
print "Shape of w", w.shape
```

```python
for j in range(1,numiter):
    J = 0
    dw = np.zeros((1,n))
    
    b =0
    a = sigmoid(np.dot(w,X)+b,play = "Forward")
   
    J = -(1.0/m) * (np.sum(np.multiply(y,np.log(a))+np.multiply(1-y,np.log(1-a))))
    #dw = (1.0/m) * np.sum(np.multiply(a-y,X),axis = 1)
    dw = (1.0/m)* np.dot((a-y),X.T)
    db = (1.0/m) * np.sum(a-y)
    w = w - learningrate * dw
    
    b = b - learningrate * db
    
    print "Loss after ",j,"iteration =>",J
    
```

#### Prediction for Traindata

```python
##Predictions
pred = []
a = sigmoid(np.dot(w,X)+b,play="Forward")

for i in range(a.shape[1]):
    if (a[0][i]>0.5):
        pred.append(1)
    elif (a[0][i] <= 0.5):
        pred.append(0)
```

```python
print "Predicted for Traindata", pred
print "Truelabel for Train data",y
```
