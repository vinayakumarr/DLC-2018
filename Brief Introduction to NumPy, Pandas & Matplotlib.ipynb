### GroupBy
* Returns a Pandas groupBy object
* List of functions applicable on groupBy object is different

```python
rental_data.groupby('Bedroom').groups
```

```python
#Mean total price of house based on number of bedrooms
gb = rental_data.groupby('Bedroom')

#Apply agg on Price col, return series of prices of all 1 bedroom, 2bedroom etc separetely
gb.Price.agg(lambda p: np.mean(p))
```

```python
def cat_house(p):
    if p < 20000:
        return 'Low'
    elif p < 150000:
        return 'Medium'
    else:
        return 'High'
    
#Map comes handy for per row per col manupulations 
rental_data['Category'] = rental_data.Price.map(cat_house)
```

### Learning More Pandas with titanic data

```python
titatic_data = pd.read_csv('data/titanic-train.csv',index_col='PassengerId')
```

```python
titatic_data.head()
```

```python
titatic_data.info()
### Conclusion
# Cabin col have lot of missing data
# Age col have few missing data

## Note: 
# Delete Cabin col
# Fill missing Age data
```

```python
#Dropping the collumn with very less useful information
#inplace - does change the dataframe itself
titatic_data.drop('Cabin',axis=1, inplace=True)
```

```python
#Find number of male or females
titatic_data.groupby('Sex').size()
```

```python
titatic_data['Embarked'] = titatic_data.Embarked.map(lambda e: 'SI' if e == 'S' else e)
```

```python
titatic_data.head()
```

```python
def f(age):
    if age < 18:
        return "kid"
    elif age >= 18:
        return "adult"
    else:
        return "not known"

#Adda new col which tell if the person is kid or adult or not known
titatic_data['Category'] = titatic_data.Age.map(f)
```

```python
titatic_data.head()
```

```python
#NameLength col added 

titatic_data['NameLength'] = titatic_data.Name.str.len()
```

```python
titatic_data.head()
```

```python
#Find the persons with longest names
max_name_length = np.max(titatic_data.NameLength)
titatic_data[titatic_data['NameLength'] == max_name_length]
```

```python
#Find rows where age is missing
titatic_data[pd.isnull(titatic_data.Age)].head()
```

```python
#Find rows where age is present
age_data = titatic_data[pd.notnull(titatic_data.Age)]
age_data.head()
```

```python
# Group passanger based on age bracket
```

```python
age_data.set_index('Age',inplace=True)
```

```python
age_data.head()
```

```python
age_data.groupby(lambda a: int(a/10)).size()
```

```python
#Finding mean of people between age group 10-39
#reset index because filtering is not possible on index col

age_data.reset_index(inplace=True)

age_df = age_data[(age_data.Age > 10) & (age_data.Age < 39)]
```

```python
age_df.Age.mean()
```

```python
#Filling missing age 
titatic_data['Age'] = titatic_data.Age.fillna(25.8)
```

```python
titatic_data.head()
```

```python
titatic_data.info()
```

```python
##Survival games
titatic_data.groupby('Survived').count()
```

```python
gbs = titatic_data.groupby('Survived')
```

```python
gbs.groups
```

```python
def f(r):
    m = np.sum(r == 'male')
    f = np.sum(r == 'female')
    return ('Male',m,'Female',f)
    
gbs.Sex.agg(f)
```
